```python
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load all .txt files from the ssot folder
source = [doc for doc in os.listdir("ssot") if doc.endswith('.txt')]
notes = [
    open(os.path.join("ssot", _file), "r", encoding="utf-8").read()
    for _file in source
]
```

```python
vectorizer = TfidfVectorizer()
```

```python
source_vectors = vectorizer.fit_transform(notes).toarray()

def similarity(doc1_vec, doc2_vec):
    """
    Compute cosine similarity between two TF-IDF vectors.
    Returns a float between 0 and 1.
    """
    return cosine_similarity(doc1_vec.reshape(1, -1), doc2_vec.reshape(1, -1))[0][0]


def create_zip(file_list, vectors):
    """
    Zip the file names with their corresponding TF-IDF vectors.
    """
    return list(zip(file_list, vectors))
```

```python
test_path = os.path.join("testing", "test1.txt")

with open(test_path, "r", encoding="utf-8") as f:
    testp_text = f.read()

test_vector = vectorizer.transform([testp_text]).toarray()[0] #test file

print(f"Test file loaded: {test_path}")

results = []
for file, vec in zip(source, source_vectors): # the ssot is the source
    sim_score = similarity(vec, test_vector)
    results.append((file, sim_score))
    print(f"{file} ↔ test1.txt : {sim_score:.4f}")
```
